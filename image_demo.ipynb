```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc("text", usetex=True)
rc("font", family = "serif")
rc("figure",figsize=(9,6))
rc("figure",facecolor="white")
%config InlineBackend.figure_format = 'retina'
from skimage.color import rgb2gray
from skimage.data import rocket, coffee
```

### Load an image and display it

```python
plates = plt.imread("plate.tif")
im = coffee()
plt.imshow(im)
plt.title(r"Array Shape {}".format(im.shape),fontsize=16)
plt.show()
print "Min pixel value: {}".format(np.min(im))
print "Max pixel value: {}".format(np.min(im))
```

So we have a 8 bit depth image - pretty standard for normal image formats (e.g. jpg, png, gif).

### Look at the color channels

```python
fig, ax = plt.subplots(1,3,figsize=(20,16))
colors = ["Reds","Greens","Blues"]
for i in range(3):
    ax[i].imshow(im[:,:,i], cmap=colors[i])
plt.show()
```

### Convert to grayscale 

Y = 0.2125 R + 0.7154 G + 0.0721 B

Apparently this is standard. Our eyes are most sensitive to green and least sensitive to blue.

```python
plt.imshow(rgb2gray(im),cmap="Greys_r")
plt.colorbar()
plt.show()
```

You'll notice that the "luminance is scaled to be between 0 and 1. Lets just re-map it to 0-255

```python
gray_cup = np.round(255*rgb2gray(im)).astype("int")
plt.imshow(gray_cup,cmap="Greys_r")
plt.colorbar()
plt.show()
print "Min pixel value: {}".format(np.min(gray_cup))
print "Max pixel value: {}".format(np.max(gray_cup))
```

```python
plt.imshow(gray_cup[180:200,320:340],cmap="Greys_r")
plt.colorbar()
plt.show()
```
